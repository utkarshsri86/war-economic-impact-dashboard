```python
#💡 Why?	Gives you a complete map of all 28 columns before diving into analysis.
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('war_economic_impact_dataset.csv')

# Shape: (rows, columns)
print('Shape:', df.shape)

# Column names and data types
print(df.dtypes)

# First 5 rows preview
df.head()
```

```python
#missing value check
#Identify any NULL or NaN values in the dataset. Missing data must be handled before ML training — it causes errors or biased models.
# Count missing values per column
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

# View as a clean table
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})
print(missing_df[missing_df['Missing Count'] > 0])

# Quick check
print('Total missing values:', df.isnull().sum().sum())

```

```python
#statistical summary
#Get mean, std, min, max, and percentiles for all numeric columns. Reveals the scale and spread of each variable at a glance.
#Helps detect outliers and extreme ranges (e.g. Cost_of_War_USD ranges from $1B to $500B).
# Summary statistics for all numeric columns
df.describe().T.round(2)

# Key columns to focus on
key_cols = [
    'GDP_Change_%', 'Inflation_Rate_%', 'Currency_Devaluation_%',
    'Cost_of_War_USD', 'Estimated_Reconstruction_Cost_USD',
    'Pre_War_Poverty_Rate_%', 'During_War_Poverty_Rate_%'
]
df[key_cols].describe().T
```

```python
#Categorical Columns Analysis
#📌 Purpose	Count the frequency of each category in text columns like Region, Conflict_Type, Status, etc. Understand class balance.
#Reveals imbalance: 45% of conflicts are in the Middle East, 36% are Civil Wars, 70% have profiteering.
cat_cols = [
    'Conflict_Type', 'Region', 'Status',
    'Most_Affected_Sector', 'Black_Market_Activity_Level',
    'War_Profiteering_Documented'
]

for col in cat_cols:
    print(f'\n=== {col} ===')
    print(df[col].value_counts())
    print(df[col].value_counts(normalize=True).mul(100).round(1))
```

```python
# target variable analysis
## Plot the distribution of Cost_of_War_USD and Estimated_Reconstruction_Cost_USD both raw and log-transformed. Skewed targets must be log-transformed for better ML performance.
#Plot the distribution of Cost_of_War_USD and Estimated_Reconstruction_Cost_USD both raw and log-transformed. Skewed targets must be log-transformed for better ML performance.
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Target Variable Distribution', fontsize=16, fontweight='bold')

# Cost of War — raw
axes[0,0].hist(df['Cost_of_War_USD']/1e9, bins=60, color='crimson', edgecolor='white')
axes[0,0].set_title('Cost of War (Billion USD) — Raw')

# Cost of War — log transformed
axes[0,1].hist(np.log1p(df['Cost_of_War_USD']), bins=60, color='darkred', edgecolor='white')
axes[0,1].set_title('Cost of War — Log Transformed')

# Reconstruction Cost — raw
axes[1,0].hist(df['Estimated_Reconstruction_Cost_USD']/1e9, bins=60,
                              color='steelblue', edgecolor='white')
axes[1,0].set_title('Reconstruction Cost (Billion USD) — Raw')

# Reconstruction Cost — log transformed
axes[1,1].hist(np.log1p(df['Estimated_Reconstruction_Cost_USD']), bins=60,
               color='navy', edgecolor='white')
axes[1,1].set_title('Reconstruction Cost — Log Transformed')

plt.tight_layout()
plt.savefig('seg5_target_distribution.png', dpi=150)
plt.show()

```
